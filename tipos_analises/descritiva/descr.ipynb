## Análise Descritiva

### Analisando Desempenho de Vendas

**Objetivo:**

* A região com o maior volume total de vendas.
* O mês em que as vendas foram mais altas.
* Visualizar as vendas totais por região em um gráfico.

Passo 1: Importando o arquivo CSV

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('vendas_regioes.csv')
```

Passo 2: Calcular a região com o maior volume total de vendas

```python
vendas_por_regiao = df.groupby('Regiao')['Vendas'].sum()
regiao_mais_vendas = vendas_por_regiao.idxmax()
vendas_max = vendas_por_regiao.max()

print(f'Região com maior volume total de vendas: {regiao_mais_vendas} ({vendas_max} vendas)')
```

Passo 3: Calcular o mês com as vendas mais altas

```python
vendas_por_mes = df.groupby('Mes')['Vendas'].sum()
mes_mais_vendas = vendas_por_mes.idxmax()
vendas_mes_max = vendas_por_mes.max()

print(f'Mês com maior volume de vendas: {mes_mais_vendas} ({vendas_mes_max} vendas)')
```

Passo 4: Visualizar vendas totais por região

```python
vendas_por_regiao.plot(kind='bar', color='#C8A0CF', figsize=(8,6))
plt.title('Vendas Totais por Região')
plt.xlabel('Região')
plt.ylabel('Total de Vendas')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
